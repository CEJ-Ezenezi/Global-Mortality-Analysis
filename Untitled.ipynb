```python
#import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
com_deaths = pd.read_csv(r"Communicable_maternal_perinatal_nutritional_conditions.csv",  sep=',', skiprows=5, index_col=False)
ill_deaths = pd.read_csv(r"Ill_defined diseases.csv",  sep=',', skiprows=5, index_col=False)
injuries = pd.read_csv(r"Injuries.csv",  sep=',', skiprows=5, index_col=False)
noncom_deaths = pd.read_csv(r"Noncommunicable diseases.csv",  sep=',', skiprows=5, index_col=False)
```

```python
#define a function that can clean up the datasets

def clean_data(dataframe):
    """deletes specified columns of the dataframe as the
    information is irrelevant and filters the years.
    parameter
    dataframe: a list of dataframes
    """
    
    df = dataframe.drop(['Region Code', 'Country Name', 
            'Age-standardized death rate per 100 000 standard population'], axis=1)
    
    return df
    
```

```python
com_deaths = clean_data(com_deaths)
```

```python
ill_deaths = clean_data(ill_deaths)
```

```python
com_deaths = com_deaths.dropna()
```

```python
ill_deaths = ill_deaths.dropna()
ill_deaths.head(50)
```

```python
injuries = injuries.dropna()
injuries.head()
```

```python
noncom_deaths = clean_data(noncom_deaths)
noncom_deaths.isna().sum()
```

```python
noncom_deaths = com_deaths.dropna()
noncom_deaths.head(50)
```

```python
# Filter the data for years between 1987 and 2021
filtered_years_data = com_deaths[(com_deaths['Year'] >= 1987) & (com_deaths['Year'] <= 2021)]
```

```python
# Filter rows where Sex is "All" and Age group code is "Age_all"
filtered_data = filtered_years_data[(filtered_years_data['Sex'] == 'All') & (filtered_years_data['Age group code'] == 'Age_all')]

# Group by Region Name and Year, then sum the Number of deaths
grouped_data = filtered_data.groupby(['Region Name', 'Year'])['Number'].sum().reset_index()

# Set the plot style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(14, 8))
sns.lineplot(data=grouped_data, x='Year', y='Number', hue='Region Name', marker='o')

# Customize the plot
plt.title('Total Deaths for All Sex and All Age Groups Over the Years by Region')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.legend(title='Region Name')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
